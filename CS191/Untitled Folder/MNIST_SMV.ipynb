```python
#Importing libraries

%matplotlib inline

import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split,GridSearchCV

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

color = sns.color_palette()
```

```python
#Loading data


df_train = pd.read_csv('train.csv',nrows=5000,skiprows = range(1,2000))

df_train.head()
```

```python
#Loading data

df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
y = df_train['label']
df_train = df_train.drop('label',axis=1)
```

```python
#Using linear boundary to separate classes

params = [
  {'C': [1, 5, 7, 10], 'random_state': [ 42, 179 ], 'kernel': ['linear']},
]   
```

```python
#Creating copies just to be safe

new_df_train = df_train.copy()
new_df_test = df_test.copy()
pca_model = PCA(n_components=80)
new_df_train = pd.DataFrame(pca_model.fit_transform(new_df_train))
new_df_test = pd.DataFrame(pca_model.transform(new_df_test))
```

```python
#Declaring Grid Search Cross Validation and SVM

gs_svm = GridSearchCV(estimator=svm.SVC(), param_grid=params, n_jobs=4)



X_train, X_valid, y_train, y_valid = train_test_split(new_df_train, y, test_size = 0.3, random_state = 42)
print(X_train.shape)
print(X_valid.shape)
print(y_train.shape)
print(y_valid.shape)
gs_svm.fit(X_train, y_train)  

print(gs_svm.best_score_)
print(gs_svm.best_params_)
y_predict = gs_svm.predict(new_df_test)  
```

```python
#Writing to file


sample_csv = pd.read_csv('sample_submission.csv')
sample_csv['Label'] = y_predict
sample_csv.to_csv("submission.csv", index=False, header=True)
```
